```python
def pre_processing(path):
    col_names = ["text", "label", "id"]
    df = pd.read_csv(path, sep="\t", names=col_names)
    
    # turn the labels into list
    df["label"] = [labels.split(",") for labels in df["label"]]

    # explod the dataframe
    df_exploded = df.explode('label')

    # change label data type
    df_exploded["label"] = df_exploded["label"].astype("int")
    
    # exclude the neutral emotion
    df_res = df_exploded[df_exploded["label"] != '0']

    return df_res
```

```python
train = pre_processing("./data/go_emotion/full_dataset/train.tsv")
```

```python
dev = pre_processing("./data/go_emotion/full_dataset/dev.tsv")
```

```python
test = pre_processing("./data/go_emotion/full_dataset/test.tsv")
```

```python
train.to_csv("./data/go_emotion/multi-label/train.csv", index=False)
```

```python
dev.to_csv("./data/go_emotion/multi-label/valid.csv", index=False)
```

```python
test.to_csv("./data/go_emotion/multi-label/test.csv", index=False)
```
