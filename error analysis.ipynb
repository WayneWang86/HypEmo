```python
import pandas as pd

# Load the data
file_path = '/content/drive/MyDrive/11777-advNLP-group-shared-drive/HW3/hypmo_output_analysis.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
error_data = data[(data['predict_label'] != data['true_label'])]

# Identifying the most mispredicted (true label, predicted label) pair
mispredicted_pairs = error_data.groupby(['true_label', 'predict_label']).size().reset_index(name='count')
most_mispredicted_pair = mispredicted_pairs.sort_values(by='count', ascending=False)

most_mispredicted_pair

```

```python
most_mispredicted_pair.head(10)
```

```python
# Selecting a few examples for each of the specified mispredicted pairs
examples_per_pair = 2  # Number of examples to select for each pair

# Defining the pairs to look for
pairs_to_examine = [
    ('anger', 'annoyance'),
    ('approval', 'admiration'),
    ('disapproval', 'annoyance'),
    ('confusion', 'curiosity'),
    ('curiosity', 'confusion'),
    ('annoyance', 'anger')
]

# Extracting and printing examples for each pair
for true_label, predict_label in pairs_to_examine:
    pair_examples = error_data[(error_data['true_label'] == true_label) &
                               (error_data['predict_label'] == predict_label)].head(examples_per_pair)

    # print(f"**Pair: ({true_label}, {predict_label})")
    for index, row in pair_examples.iterrows():
        print(f"- True Label: {row['true_label']}")
        print(f"- Predicted Label: {row['predict_label']}")
        print(f"- Text: {row['text']}\n")
```
