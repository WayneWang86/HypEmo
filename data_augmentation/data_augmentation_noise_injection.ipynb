```python
import pandas as pd
import random
```

```python
train_data = pd.read_csv('/content/HypEmo/data/go_emotion/train.csv')
```

```python
label2idx = {'admiration': 0, 'amusement': 1, 'anger': 2,
      'annoyance': 3, 'approval': 4, 'caring': 5,
      'confusion': 6, 'curiosity': 7, 'desire': 8,
      'disappointment': 9, 'disapproval': 10, 'disgust': 11,
      'embarrassment': 12, 'excitement': 13, 'fear': 14,
      'gratitude': 15, 'grief': 16, 'joy': 17,
      'love': 18, 'nervousness': 19, 'optimism': 20,
      'pride': 21, 'realization': 22, 'relief': 23,
      'remorse': 24, 'sadness': 25, 'surprise': 26}
idx2label = {}
for key, val in label2idx.items():
    idx2label[val] = key
train_data['true_label'] = train_data['label'].replace(idx2label)

train_data_sub = train_data[train_data['true_label'].isin(['annoyance', 'approval', 'caring', 'confusion', 'desire', 'disappointment', 'disapproval', 'excitement', 'nervousness', 'grief', 'pride', 'realization', 'relief'])]
```

```python
sentence = "I love this natural language processing class"
sentence_length = len(sentence)

# Define the keyboard mapping with letters and numbers
keyboard_mapping = {
    'a': 'qwsz',
    'b': 'vfghn',
    'c': 'xdfv',
    'd': 'serfcx',
    'e': 'wrd123',
    'f': 'drtgvc',
    'g': 'ftyhbv',
    'h': 'gyujnb',
    'i': 'ujo456',
    'j': 'huiknm',
    'k': 'jiolm',
    'l': 'kop',
    'm': 'njk',
    'n': 'bhjm',
    'o': 'iklp789',
    'p': 'ol0',
    'q': 'aw12',
    'r': 'etdf345',
    's': 'azxwe',
    't': 'ryfg67',
    'u': 'yihj8',
    'v': 'cfgb',
    'w': 'qase23',
    'x': 'zasdc',
    'y': 'tghu',
    'z': 'asx1',
    '0': 'pl9',
    '1': 'qaz2',
    '2': 'qw13',
    '3': 'we24',
    '4': 'er35',
    '5': 'rt46',
    '6': 'ty57',
    '7': 'yu68',
    '8': 'ui79',
    '9': 'io80'
}

def augment_text(text):
    # Convert the text to a list of characters
    char_list = list(text)

    # Generate a random index to select a character
    random_index = random.randint(0, len(char_list) - 1)
    selected_char = char_list[random_index]

    # Check if the selected character is alphanumeric
    if selected_char.isalnum():
        # Get the neighbors from the keyboard mapping
        neighbors = keyboard_mapping.get(selected_char.lower(), '')

        # Check if there are neighbors available
        if neighbors:
            # Select a random neighbor character
            replacement_char = random.choice(neighbors)

            # Replace the selected character with the replacement
            char_list[random_index] = replacement_char

    # Convert the list of characters back to a string
    augmented_text = ''.join(char_list)

    return augmented_text


def noise_augmentation(sentence):
    original_set = sentence
    sentence_length = len(sentence)
    max_add = int(min(1, len(sentence.split()) / 10))

    random_number = random.randint(1, max_add + 1)
    for _ in range(random_number):
        sentence = augment_text(sentence)
    while sentence == original_set:
        sentence = augment_text(sentence)
    # print(random_number)
    return sentence
```

```python
df = train_data_sub.reset_index(drop = True)
original_text = df['text']
df['text'] = df['text'].apply(noise_augmentation)
df = df[~df['text'].isin(original_text)]
# df[df['text'].isin(original_text)]
```
