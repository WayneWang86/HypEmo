```python
lst = [
    a + 'noise_augmentation.csv',
    a + 'gpt_augmentation.csv',
    a + 'word2_augmentation.csv',
    a + 'emoji_augmentation'
]
```

```python
noise = pd.read_csv(a + 'noise_augmentation.csv')
gpt = pd.read_csv(a + 'gpt_augmentation.csv')
word2 = pd.read_csv(a + 'word2_augmentation.csv')
emoji = pd.read_csv(a + 'emoji_augmentation.csv')
```

```python
noise = noise[noise['true_label'].isin(["caring","confusion","excitement","nervousness","pride","relief"])]
```

```python
gpt = gpt[gpt['true_label'].isin(["disapproval"])]
```

```python
word2 = word2[word2['true_label'].isin(['annoyance'])]
```

```python
emoji = emoji[emoji['true_label'].isin(['approval', 'disappointment'])]
```

```python
df = pd.concat([noise, gpt, word2, emoji], ignore_index=True)
```

```python
df.true_label.value_counts()
```

```python
df =  df.sample(frac=1).reset_index(drop=True)
```
